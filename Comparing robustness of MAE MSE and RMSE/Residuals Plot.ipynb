# Creating image to show the residuals

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Creating the Dataset

```python
size = 25
```

```python
# Creating a synthetic dataset
x = np.random.uniform(0, 10, size)
noise = np.random.normal(0, 2, size)
y = 1.14 * x + 1.75 +  noise
```

```python
# Show us the data
sns.scatterplot(x = x, y = y)
```

```python
df = pd.DataFrame()
df['x'] = x
df['y'] = y
```

## Fitting a linear regression

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(x.reshape(-1, 1), y)
```

```python
# Getting the intercept
model.intercept_
```

```python
# Getting the only coefficient
model.coef_
```

## Plotting the regression

```python
# Plots the data points and the regression
fig = plt.figure(figsize = (10,5))
ax = plt.axes()
sns.scatterplot(x = x, y = y, ax = ax)
sns.lineplot(x = [0, 10], y = [model.intercept_, (10 * model.coef_[0] + model.intercept_)], ax = ax, color = 'b')

# Plots the residuals
for i, j in zip(x, y):
    yreg = i * model.coef_[0] + model.intercept_
    ax.plot([i, i], [j, yreg], color = 'r')
```

```python
# Saves the figure
fig.savefig('residuals.jpg')
```
