## Imports

```python
from math import sin
import seaborn as sns
import matplotlib.pyplot as plt
```

## Create base functions

```python
def x1(t):
    return sin(t)
```

```python
def x2(t):
    return 2*t+3
```

```python
def x3(t):
    return 2*(t**2.0) + 3*t + 1
```

```python
def x4(t):
    return 1 if int(t)%2 == 0 else -1
```

```python
T = [t/100 for t in range(2000)]
```

```python
X1 = [x1(t) for t in T]
X2 = [x2(t) for t in T]
X3 = [x3(t) for t in T]
X4 = [x4(t) for t in T]
```

```python
plt.figure(figsize = (20,4))
ax1 = plt.subplot(1, 4, 1)
ax2 = plt.subplot(1, 4, 2)
ax3 = plt.subplot(1, 4, 3)
ax4 = plt.subplot(1, 4, 4)
sns.lineplot(x = T, y = X1, ax = ax1)
sns.lineplot(x = T, y = X2, ax = ax2)
sns.lineplot(x = T, y = X3, ax = ax3)
sns.lineplot(x = T, y = X4, ax = ax4)
ax1.set_title("Sine function")
ax2.set_title("Linear function")
ax3.set_title("Quadratic function")
ax4.set_title("Square signal")
```

## Convolution function

```python
def discrete_conv(x, w, t):
    s = 0
    for a in range(-1000, 1000):
        s += x(a) * w(t-a)
    return s
```

```python
A = [a/10 for a in range(-20, 80)]
```

## Constant kenel

```python
def w1(a):
    return 1 if a >= 1 and a <= 3 else 0
```

```python
W1 = [w1(a) for a in A]
```

```python
S1 = [discrete_conv(x1, w1, t) for t in T]
S2 = [discrete_conv(x2, w1, t) for t in T]
S3 = [discrete_conv(x3, w1, t) for t in T]
S4 = [discrete_conv(x4, w1, t) for t in T]
```

```python
fig = plt.figure(figsize = (20,4))
ax0 = plt.subplot(1, 5, 1)
ax1 = plt.subplot(1, 5, 2)
ax2 = plt.subplot(1, 5, 3)
ax3 = plt.subplot(1, 5, 4)
ax4 = plt.subplot(1, 5, 5)
sns.lineplot(x = A, y = W1, color = 'r', ax = ax0)
sns.lineplot(x = T, y = X1, ax = ax1)
sns.lineplot(x = T, y = X2, ax = ax2)
sns.lineplot(x = T, y = X3, ax = ax3)
sns.lineplot(x = T, y = X4, ax = ax4)
sns.lineplot(x = T, y = S1, ax = ax1, color = 'g')
sns.lineplot(x = T, y = S2, ax = ax2, color = 'g')
sns.lineplot(x = T, y = S3, ax = ax3, color = 'g')
sns.lineplot(x = T, y = S4, ax = ax4, color = 'g')
fig.suptitle("Constant Kernel", fontsize=20)
ax0.set_title("Kernel")
ax1.set_title("Sine function")
ax2.set_title("Linear function")
ax3.set_title("Quadratic function")
ax4.set_title("Square signal")
plt.tight_layout()
```

## Linear kenel

```python
def w2(a):
    return a/3 if a >= 1 and a <= 3 else 0
```

```python
W2 = [w2(a) for a in A]
```

```python
S1 = [discrete_conv(x1, w2, t) for t in T]
S2 = [discrete_conv(x2, w2, t) for t in T]
S3 = [discrete_conv(x3, w2, t) for t in T]
S4 = [discrete_conv(x4, w2, t) for t in T]
```

```python
fig = plt.figure(figsize = (20,4))
ax0 = plt.subplot(1, 5, 1)
ax1 = plt.subplot(1, 5, 2)
ax2 = plt.subplot(1, 5, 3)
ax3 = plt.subplot(1, 5, 4)
ax4 = plt.subplot(1, 5, 5)
sns.lineplot(x = A, y = W2, color = 'r', ax = ax0)
sns.lineplot(x = T, y = X1, ax = ax1)
sns.lineplot(x = T, y = X2, ax = ax2)
sns.lineplot(x = T, y = X3, ax = ax3)
sns.lineplot(x = T, y = X4, ax = ax4)
sns.lineplot(x = T, y = S1, ax = ax1, color = 'g')
sns.lineplot(x = T, y = S2, ax = ax2, color = 'g')
sns.lineplot(x = T, y = S3, ax = ax3, color = 'g')
sns.lineplot(x = T, y = S4, ax = ax4, color = 'g')
fig.suptitle("Linear Kernel", fontsize=20)
ax0.set_title("Kernel")
ax1.set_title("Sine function")
ax2.set_title("Linear function")
ax3.set_title("Quadratic function")
ax4.set_title("Square signal")
plt.tight_layout()
```

## Quadratic kenel

```python
def w3(a):
    return ((a-1)**2)/4 if a >= 1 and a <= 3 else 0
```

```python
W3 = [w3(a) for a in A]
```

```python
S1 = [discrete_conv(x1, w3, t) for t in T]
S2 = [discrete_conv(x2, w3, t) for t in T]
S3 = [discrete_conv(x3, w3, t) for t in T]
S4 = [discrete_conv(x4, w3, t) for t in T]
```

```python
fig = plt.figure(figsize = (20,4))
ax0 = plt.subplot(1, 5, 1)
ax1 = plt.subplot(1, 5, 2)
ax2 = plt.subplot(1, 5, 3)
ax3 = plt.subplot(1, 5, 4)
ax4 = plt.subplot(1, 5, 5)
sns.lineplot(x = A, y = W3, color = 'r', ax = ax0)
sns.lineplot(x = T, y = X1, ax = ax1)
sns.lineplot(x = T, y = X2, ax = ax2)
sns.lineplot(x = T, y = X3, ax = ax3)
sns.lineplot(x = T, y = X4, ax = ax4)
sns.lineplot(x = T, y = S1, ax = ax1, color = 'g')
sns.lineplot(x = T, y = S2, ax = ax2, color = 'g')
sns.lineplot(x = T, y = S3, ax = ax3, color = 'g')
sns.lineplot(x = T, y = S4, ax = ax4, color = 'g')
fig.suptitle("Quadratic Kernel", fontsize=20)
ax0.set_title("Kernel")
ax1.set_title("Sine function")
ax2.set_title("Linear function")
ax3.set_title("Quadratic function")
ax4.set_title("Square signal")
plt.tight_layout()
```
