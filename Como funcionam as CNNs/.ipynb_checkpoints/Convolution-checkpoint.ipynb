```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = ""
```

```python
# Montando drive (colaboratory)
from google.colab import drive
drive.mount('/content/drive')
path += '/content/drive/My Drive/Colab Notebooks/'
```

```python
image = cv2.imread(path + 'BruceSquare4.png')
```

```python
kernel_sharpen = np.array([[0, -1, 0],
                           [-1, 5, -1],
                           [0, -1, 0]])
```

```python
sharpened = cv2.filter2D(src=image, ddepth=-1, kernel=kernel_sharpen)
```

```python
kernel_blur = np.array([[1, 1, 1],
                       [1, 1, 1],
                       [1, 1, 1]])
kernel_blur = 1/9 * kernel_blur
```

```python
blurred = cv2.filter2D(src=image, ddepth=-1, kernel=kernel_blur)
```

```python
fig, ax = plt.subplots(1,3,figsize=(15,9))
ax[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
ax[1].imshow(cv2.cvtColor(sharpened, cv2.COLOR_BGR2RGB))
ax[2].imshow(cv2.cvtColor(blurred, cv2.COLOR_BGR2RGB))
ax[0].set_title("Original (a)")
ax[1].set_title("Sharpen (b)")
ax[2].set_title("Blur (c)")
ax[0].axis('off')
ax[1].axis('off')
ax[2].axis('off')
fig.show()
```

### Aplicando os filtros

```python
input_image = image
```

```python
gray = cv2.cvtColor(input_image, cv2.COLOR_BGR2GRAY)
```

```python
kernel_v = np.array([[1, 0, -1],
                     [1, 0, -1],
                     [1, 0, -1]])
```

```python
out_v = cv2.filter2D(src=gray, ddepth=-1, kernel=kernel_v)
```

```python
kernel_h = np.array([[1, 1, 1],
                     [0, 0, 0],
                     [-1, -1, -1]])
```

```python
out_h = cv2.filter2D(src=gray, ddepth=-1, kernel=kernel_h)
```

```python
fig, ax = plt.subplots(1,3,figsize=(15,9))
ax[0].imshow(gray, cmap = 'gray')
ax[1].imshow(out_v, cmap = 'gray')
ax[2].imshow(out_h, cmap = 'gray')
ax[0].set_title("Escala de Cinza (d)")
ax[1].set_title("Filtro Vertical (e)")
ax[2].set_title("Filtro Horizontal (f)")
ax[0].axis('off')
ax[1].axis('off')
ax[2].axis('off')
fig.show()
```

### Padding

```python
padding = 10
```

```python
import torch
from torch import nn
import torchvision
from torchvision.io import read_image
```

```python
tensor = torch.Tensor(image)
tensor.shape
```

```python
tensor = tensor.permute(2, 0, 1)
tensor.shape
```

```python
tensor = torch.unsqueeze(tensor, 0)
tensor.shape
```

```python
reflect = nn.ReflectionPad2d(padding)
reflected_img = reflect(tensor)
reflected_img.shape
```

```python
reflected_img = reflected_img[0].permute(1, 2, 0)
reflected_img = reflected_img.numpy()
reflected_img = reflected_img.astype('uint8')
reflected_img = cv2.cvtColor(reflected_img, cv2.COLOR_BGR2RGB)
reflected_img.shape
```

```python
zeropad = nn.ZeroPad2d(padding)
zeropadded_img = zeropad(tensor)
zeropadded_img.shape
```

```python
zeropadded_img = zeropadded_img[0].permute(1, 2, 0)
zeropadded_img = zeropadded_img.numpy()
zeropadded_img = zeropadded_img.astype('uint8')
zeropadded_img = cv2.cvtColor(zeropadded_img, cv2.COLOR_BGR2RGB)
zeropadded_img.shape
```

```python
constantpad = nn.ConstantPad2d(padding, value = 155)
constpadded_img = constantpad(tensor)
constpadded_img.shape
```

```python
constpadded_img = constpadded_img[0].permute(1, 2, 0)
constpadded_img = constpadded_img.numpy()
constpadded_img = constpadded_img.astype('uint8')
constpadded_img = cv2.cvtColor(constpadded_img, cv2.COLOR_BGR2RGB)
constpadded_img.shape
```

```python
fig, ax = plt.subplots(1,3,figsize=(15,9))
ax[0].imshow(zeropadded_img)
ax[1].imshow(reflected_img)
ax[2].imshow(constpadded_img)
ax[0].set_title("Zero Padding (a)")
ax[1].set_title("Reflection Padding (b)")
ax[2].set_title("Constant Padding (c)")
ax[0].axis('off')
ax[1].axis('off')
ax[2].axis('off')
fig.show()
```
